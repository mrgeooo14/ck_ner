```python
import pandas as pd
import ast

ck_df = pd.read_csv('./data/presto_max_as_csv.csv', index_col = 0)
ck_df['ent'] = ck_df['ent'].apply(ast.literal_eval)
ck_df['tokens'] = ck_df['tokens'].apply(ast.literal_eval)
ck_df
```

```python
mapping_dict = {
    "pers": "PERSON",
    "loc": "LOCATION",
    "amount": "AMOUNT",
    "func": "FUNCTION",  
    "time": "TIME",
    "org": "ORGANIZATION",
    "event": "EVENT",
    "prod": "PRODUCT",
}

#### Apply a lambda function to convert each entity tag to the mapping above
ck_df['ent'] = ck_df['ent'].apply(
    lambda tag_list: [
        "O" if tag == "O" 
        else mapping_dict.get(tag.split('-')[-1].lower(), tag.split('-')[-1].upper())
        for tag in tag_list
    ]
)
```

Expected training data is : *training_data = [
  ("Tokyo Tower is 333m tall.", [(0, 11, "BUILDING")]),
]*, a list of tuples (sentence, entities)

Hence, the process below converts it into it

```python
training_data = []

for idx, row in ck_df.iterrows():
    text = row["text"]
    tokens = row["tokens"]
    labels = row["ent"]

    spans = []            ### Will store the (start, end, label) tuples for entities
    current_span = None   ### Active span: (start, end, label)
    pointer = 0           ### Character pointer in the sentence to find the char id's of the entities

    for token, label in zip(tokens, labels):
        token = token.strip()
        while pointer < len(text) and text[pointer].isspace():
            pointer += 1

        #### Pointer loops through tokens
        token_len = len(token)
        ### Token found, pointer hops onto the next
        if text[pointer:pointer+token_len] == token:
            start = pointer
            end = start + token_len
            pointer = end 
        else:
            start = text.find(token, pointer)
            if start == -1:
                continue
            end = start + token_len
            pointer = end 

        #### If the found token is an entity (different from 'O' -> Other), start the process
        if label != "O":
            ### Merge Entity Spans separated by a whitespace
            if current_span is not None and current_span[2] == label:
                gap = text[current_span[1]:start]
                if gap.strip() == "":
                    current_span = (current_span[0], end, label)
                else:
                    ### Merge Entity Spans separated by something else
                    spans.append(current_span)
                    current_span = (start, end, label)
            else:
                if current_span is not None:
                    spans.append(current_span)
                current_span = (start, end, label)
        else:
            if current_span is not None:
                spans.append(current_span)
                current_span = None

    if current_span is not None:
        spans.append(current_span)
    training_data.append((text, spans))
```

Convert Train Set to spacy format

```python
import spacy
from spacy.tokens import DocBin

nlp = spacy.blank("fr")

print("DEBUG")
errors = 0
db = DocBin()
for text, annotations in dev_data:
    doc = nlp(text)
    ents = []
    for start, end, label in annotations:
        span = doc.char_span(start, end, label=label , alignment_mode="expand")
        if span is None:
            # Print a warning or handle the misaligned entity as needed.
            print('ciao, ', annotations)
            print('ok, ', text)
            print(f"Skipping entity: {start}-{end} {label} (could not be aligned)")
        else:
            ents.append(span)
    doc.ents = ents  # set doc.ents to the list of valid spans (even if empty)
    # print("S: ", text)
    if ents == [] and annotations != []:
        errors += 1
    
    db.add(doc)
# db.to_disk("./dev_ck.spacy")
```
